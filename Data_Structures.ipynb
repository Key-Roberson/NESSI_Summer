```python
ds["precipitation"] = (("x", "y", "time"), precip)
```

```python
ds.coords["lat"] = (("x", "y"), lat)
```

```python
ds.coords["lon"] = (("x", "y"), lon)
```

```python
ds.coords["time"] = pd.date_range("2014-09-06", periods=3)
```

```python
ds.coords["reference_time"] = pd.Timestamp("2014-09-05")
```
