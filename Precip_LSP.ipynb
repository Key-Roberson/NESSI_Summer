```python
import xarray as xr
import glob
```

```python
firstlat = 15
secondlat = 60
firstlon = -130+360
secondlon = -60+360
```

```python
lsp_list = glob.glob("/glade/scratch/molina/cesm_mcs/e5.*_lsp.*.nc")
```

```python
lsp_list = sorted(lsp_list)
```

```python
for i, precip in enumerate(lsp_list): 
    
    if (i % 2) != 0:
        
        if i==1:
            ds1 = xr.open_dataset(lsp_list[i-1])["LSP"]
            ds1 = ds1.where((ds1["longitude"]>=firstlon)&(ds1["longitude"]<=secondlon)&(ds1["latitude"]>=firstlat)&(ds1["latitude"]<=secondlat),drop=True)
            ds2 = xr.open_dataset(precip)["LSP"][:,:,:-6]
            ds2 = ds2.where((ds2["longitude"]>=firstlon)&(ds2["longitude"]<=secondlon)&(ds2["latitude"]>=firstlat)&(ds2["latitude"]<=secondlat),drop=True)
            ds = xr.concat([ds1, ds2],dim="time")
            ds = ds.resample(time='MS').sum(skipna=True)
            ds.to_netcdf("ERA5_LSP_"+ds.coords["time"].dt.strftime("%Y").values[0]+"_"+ds.coords["time"].dt.strftime("%m").values[0]+".nc")


        if i > 1:
            ds1 = xr.open_dataset(lsp_list[i-2])["LSP"][:,:,-6:]
            ds1 = ds1.where((ds1["longitude"]>=firstlon)&(ds1["longitude"]<=secondlon)&(ds1["latitude"]>=firstlat)&(ds1["latitude"]<=secondlat),drop=True)
            ds2 = xr.open_dataset(lsp_list[i-1])["LSP"]
            ds2 = ds2.where((ds2["longitude"]>=firstlon)&(ds2["longitude"]<=secondlon)&(ds2["latitude"]>=firstlat)&(ds2["latitude"]<=secondlat),drop=True)
            ds3 = xr.open_dataset(precip)["LSP"][:,:,:-6]
            ds3 = ds3.where((ds3["longitude"]>=firstlon)&(ds3["longitude"]<=secondlon)&(ds3["latitude"]>=firstlat)&(ds3["latitude"]<=secondlat),drop=True)
            ds = xr.concat([ds1, ds2, ds3],dim="time")
            ds = ds.resample(time='MS').sum(skipna=True)
            ds.to_netcdf("ERA5_LSP_"+ds.coords["time"].dt.strftime("%Y").values[0]+"_"+ds.coords["time"].dt.strftime("%m").values[0]+".nc")
            
        print(i, precip)
```
