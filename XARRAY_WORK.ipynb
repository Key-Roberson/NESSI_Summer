# Nearest neighbor lookup

```python
#How to find the nearest neighbor to the listed value using nearest command
ds["VAR_2T"].sel(latitude=[1.1, 1.9], method='nearest')
```

```python
#to find the nearest neighbor to the listed value using backfill command
ds["VAR_2T"].sel(latitude=0.5, method="backfill")
```

```python
#To find the nearest neighbor to the listed values using pad command
ds["VAR_2T"].reindex(latitude=[0.5, 1, 1.5, 2, 2.5], method="pad")
```

```python
#To give a maximum deviating decimal when using an inexact lookup method using the tolerance command
ds["VAR_2T"].reindex(latitude=[1.1, 1.5], method="nearest", tolerance=0.2)
```

```python
#to print all the values in the selected varible that are exact and close in the range given
ds["VAR_2T"].sel(longitude=slice(0.9,3.1))
```
